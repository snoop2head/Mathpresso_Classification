```python
from konlpy.tag import Okt

def delete_josa_okt(raw:str, stopwords:list) -> str:
    okt_tokenizer = Okt()
    # position arguments(or position tags): 내가 뽑아내고 싶은 형태소들
    okt_pos = ["Josa"]
    
    # norm: normalize 그랰ㅋㅋ -> 그래ㅋㅋ  # stem: stemming 바뀌나->바뀌다
    okt_list = [word for word, tag in okt_tokenizer.pos(raw, norm=False, stem=False)
        # tag에 속하지 않고, stopwords에 속하지 않는 경우
        if tag not in okt_pos and word not in stopwords 
    ]
    str_text = " ".join(okt_list)

    return str_text
```

```python
import re
```

```python
# Math symbols
math_terms = ['sin', 'cos', 'tan', #Triangle function
             'alpha', 'beta', 'theta', 'gamma', 'omega', 'phi', #Roman letters
             'uu', 'nn', 'sup', 'sub', '^C', #Sets
             'pi', 'abs', 'sqrt', 'cdots', #Etc
             '!', '(', ')', '{', '}', '[', ']', '+', '|', '/_', '@', '=' 
             ]

# 좌표 괄호와 일반 괄호 구분 : 추가 예정

# Math symbols which need regular expression
math_terms_complex = [r'_[0-9]+C_[0-9+]', #Combination
                     r'_[0-9]+P_[0-9+]' #Permutation
                     ] 

# Symbols which are not related to math
non_math_terms = ['<br/>', '<fieldset>', '<legend>', '</fieldset>', '</legend>', '\{::\}']
```

```python
df = pd.read_csv(f'{DATA_PATH}/몽데이크_Open.csv')
```

```python
# Add spaces to the next of math symbols
for i in math_terms:
    df['text'] = df['text'].apply(lambda x: x.replace(i, ' '+i+' '))

# Remove non-math terms
for i in non_math_terms:
    df['text'] = df['text'].apply(lambda x: x.replace(i, ' '))

# Handle exceptions
df['text'] = df['text'].apply(lambda x: re.sub(r'_[0-9]+C_[0-9+]', 'Combination', x))
df['text'] = df['text'].apply(lambda x: re.sub(r'_[0-9]+P_[0-9+]', 'Permutation', x))
df['text'] = df['text'].apply(lambda x: re.sub(r'<보기>', '보기', x))
```

```python
def strip_josa(list_input:list, list_stopword:list) -> list:
    
    # 이미 단어 단위로 분리되어 있는 조사들을 제거
    list_filtered = [item for item in list_input if item not in list_stopword]

    # 아직 단어에 붙어 있는 다섯 글자 짜리 조사와 어미들을 제거
    list_five_filtered = []

    for item_five in list_filtered:
        if len(item_five) <=5:
            list_five_filtered.append(item_five)
        elif len(item_five)>=6 and item_five[-5:] not in list_stopword:
            list_five_filtered.append(item_five)
        elif len(item_five)>=6 and item_five[-5:] in list_stopword:
            list_five_filtered.append(item_five[:-5])

    # 아직 단어에 붙어 있는 네 글자 짜리 조사와 어미들을 제거
    list_four_filtered = []

    for item_four in list_five_filtered:
        if len(item_four) <=4:
            list_four_filtered.append(item_four)
        elif len(item_four)>=5 and item_four[-4:] not in list_stopword:
            list_four_filtered.append(item_four)
        elif len(item_four)>=5 and item_four[-4:] in list_stopword:
            list_four_filtered.append(item_four[:-4])

    # 아직 단어에 붙어 있는 세 글자 짜리 조사와 어미들을 제거
    list_three_filtered = []

    for item_three in list_four_filtered:
        if len(item_three) <=3:
            list_three_filtered.append(item_three)
        elif len(item_three)>=4 and item_three[-3:] not in list_stopword:
            list_three_filtered.append(item_three)
        elif len(item_three)>=4 and item_three[-3:] in list_stopword:
            list_three_filtered.append(item_three[:-3])
    
    # 아직 단어에 붙어 있는 두 글자 짜리 조사와 어미들을 제거
    list_two_filtered = []

    for item_two in list_three_filtered:
        if len(item_two) <=2:
            list_two_filtered.append(item_two)
        elif len(item_two)>=3 and item_two[-2:] not in list_stopword:
            list_two_filtered.append(item_two)
        elif len(item_two)>=3 and item_two[-2:] in list_stopword:
            list_two_filtered.append(item_two[:-2])
    

    # 아직 단어에 붙어 있는 한 글자 짜리 조사와 어미들을 제거
    list_one_filtered = []
    for item in list_two_filtered:
        if len(item) <=1:
            list_one_filtered.append(item)
        elif len(item)>=2 and item[-1] not in list_stopword:
            list_one_filtered.append(item)
        elif len(item)>=2 and item[-1] in list_stopword:
            list_one_filtered.append(item[:-1])
    
    
    list_output = list_one_filtered

    return list_output
```
